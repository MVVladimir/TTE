# Собираем refs из ways

```python
import pandas as pd
```

Собираем данные по тегам

```python
from xml.dom import minidom

# parse an xml file by name
file = minidom.parse('../datasets/map.xml')

#use getElementsByTagName() to get tag
ways = file.getElementsByTagName('way')
nodes = file.getElementsByTagName('node')
```

Создаём dict

```python
refsAndIds = {}

for elem in ways:
    err = False
    refsOne = []
    id = elem.getAttribute('id')
    nodesListND = elem.getElementsByTagName('nd')
    nodeListK = elem.getElementsByTagName('tag')
    for nodeK in nodeListK:
        k = nodeK.attributes.get('k')
        if (k != 'highway'):
            err = True
    if (err == True):
        continue
    for nodeND in nodesListND:
        ref = nodeND.attributes.get('ref')
        refsOne.append(ref.value)
    refsAndIds.update({id: refsOne})
```

Сколько всего точек

```python
setOfNodes = []
for key in refsAndIds.keys():
    vals = refsAndIds.values()
    print
    print()
    for val in vals:
     setOfNodes.append(val)
```
